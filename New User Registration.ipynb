# Registering New User

```python
import os
import cv2
import uuid
import numpy as np
import tensorflow as tf
from mtcnn.mtcnn import MTCNN
```

```python
def save_face(img_path, dest_folder):
    img = cv2.imread(img_path)
    detector = MTCNN()
    faces = detector.detect_faces(img)

    #fetching the (x,y)co-ordinate and (width-->w, height-->h) of the image
    try:
        x1,y1,w,h = faces[0]['box']
        x1, y1 = abs(x1), abs(y1)
        x2 = abs(x1+w)
        y2 = abs(y1+h)

        #locate the co-ordinates of face in the image
        store_face = img[y1:y2,x1:x2]
        store_face = cv2.resize(store_face, (224, 224)) #The VGGFace model expects a 224x224x3 size face image as input, and it outputs a face embedding vector with a length of 2048.
        cv2.imwrite(os.path.join(dest_folder, os.path.basename(img_path)), store_face)
        return "Face Found"
    except IndexError:
        return "Out of Index"
```

```python
def preprocess(file_path):
    byte_img = tf.io.read_file(file_path)
    img = tf.io.decode_jpeg(byte_img)
    return img
```

## Making New Directory

```python
new_ID = str(input("Enter New ID: "))
```

```python
try:
    path = os.path.join('data', 'verifications', new_ID)
    os.makedirs(path)
except OSError:
    print("User already registered")
```

## Capturing verification images

```python
path = os.path.join('data', 'verifications', new_ID)

cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('Registration', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('v'):
        imgname = os.path.join(path, '{}.jpg'.format(uuid.uuid1()))
        cv2.imwrite(imgname, frame)
        
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

## Detecting and cropping out faces

```python
path = os.path.join('data', 'verifications', new_ID)

os.makedirs(os.path.join('data', 'verifications', new_ID, "faces"))

for face in os.listdir(path):
    if len(os.listdir(os.path.join(path, "faces"))) < 50:
        save_face(os.path.join(path, face), os.path.join(path, "faces"))
```
